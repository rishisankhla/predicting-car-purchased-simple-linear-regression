```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('mtcars_for_manymerge.csv')
df
```

```python
x = df['hp']
y = df['cyl']
x
y
```

```python
fig, ax = plt.subplots()
fit = np.polyfit(x, y, deg=1)
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)
fig.show()
```

```python
sns.regplot(data=df,x='hp',y='cyl')
plt.show()
```
